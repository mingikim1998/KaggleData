```python
# categorize income groups into 6
import numpy as np
housing["income_cat"] = pd.cut(housing["median_income"],
                              bins=[0, 1, 3, 4, 5, 6, 15.1],
                              labels = [1,2,3,4,5,6]) # 표본편차 보고 정함
housing['income_cat'].hist()
```

```python
housing["income_cat"] = pd.cut(housing["median_income"],
                              bins=[0, 3, 4, 5, 6, np.inf], # bin 최소값 지정하지 않으면 NaN 으로 나옴
                              labels = [2,3,4,5,6]) # 표본편차 보고 정함
housing['income_cat'].hist()
```

```python
housing["income_cat"] = pd.cut(housing["median_income"],
                              bins=[2., 3., 4., 5., 6., np.inf],
                              labels = [2,3,4,5,6]) # 표본편차 보고 정함
housing['income_cat'].hist()
```

```python
housing['income_cat']
```

```python
# stratifed Shuffle Split *********
# 시행착오 ValueError: Input y contains NaN. -> NaN 값 없애주고 돌려야함
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, test_size=0.2, random_state= 42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]

strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

```python
strat_train_set["income_cat"].value_counts() / len(strat_test_set) # 복습할 부분
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis =1, inplace=True)

# If False, return a copy. Otherwise, do operation inplace and return None.
```

```python
housing = strat_train_set.copy()
housing.plot(kind="scatter", x="longitude", y="latitude")
```
